# (10)- Identify columns that should be numerical but contain nulls or inconsistencies.

```python
df.describe(include='int64')


import pandas as pd

# 1️⃣ View data types to spot numeric columns stored incorrectly (e.g., as object)
df.dtypes


# 2️⃣ Check missing values in numeric columns
# These columns SHOULD be numeric but already have nulls
numeric_nulls = df.select_dtypes(include='number').isna().sum()
numeric_nulls[numeric_nulls > 0]


# 3️⃣ Identify object columns that are mostly numeric
# These likely contain numbers mixed with strings like 'N/A', '-', etc.
object_cols = df.select_dtypes(include='object').columns

numeric_like = df[object_cols].apply(
    lambda col: pd.to_numeric(col, errors='coerce').notna().mean()
)

# Columns with >70% numeric values are strong candidates
numeric_like[numeric_like > 0.7]


# 4️⃣ Combine results to get final problematic columns
problematic_numeric_columns = list(
    set(numeric_nulls[numeric_nulls > 0].index)
    | set(numeric_like[numeric_like > 0.7].index)
)

problematic_numeric_columns


# 5️⃣ Quick statistical sanity check
# Look for odd min/max or missing counts
df[problematic_numeric_columns].describe(include='all')
```

# (12) -Standardize the gender column (M/F → Male/Female).



```python
df['gender'].unique()
df['gender'] = df['gender'].replace({
    'M': 'Male',
    'Male': 'Male',
    'F': 'Female',
    'Female': 'Female'
})

# Even better (handles lowercase too)

df['gender'] = (
    df['gender']
    .str.strip()
    .str.upper()
    .replace({'M': 'Male', 'F': 'Female'})
)
```

# (14) -Handle missing values in quantity.



```python
df.head(3)

# -- df['quantity'].isna().sum()
#  - df.fillna({'quantity' : df['quantity'].mean()})

# backward fill
df['quantity'] = df['quantity'].bfill()
df['quantity'].isna().sum()
```

# (17) - Replace null city values logically based on country (if possible).



```python
# df['city'].isna().sum()
city_by_country = (
    df.dropna(subset=['city'])
      .groupby('country')['city']
      .agg(lambda x: x.mode()[0])
)
df['city'] = df['city'].fillna(df['country'].map(city_by_country))
df['city'].isna().sum()
```

# (27) - Create a discount_amount column.



```python
#Safer version (handles missing discounts)
df['discount_amount'] = df['total_amount'] * (1 - df['discount'].fillna(0) / 100)

# if discount is a percentage
df['discount_amount'] = df['total_amount'] - (df['total_amount'] * df['discount'] / 100)
df
```
